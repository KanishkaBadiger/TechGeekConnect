```python
# 1.] Accept any 5 digit number from user and print 
# addition of its individual digits
# (Example: input no.=12345 , output = 15)


num = input("Enter any 5 numbers: ")
sum=0
for i in num:
    sum+=int(i)

print("Sum of individual digits is:", sum)
```

```python
# 3.] Basic salary is input to the keyboard. 
# Write a program to print a gross salary 
# if da is 58% of basic, ta is 30% of basic and hra is 38% of basic.

basic_salary = float(input("Enter basic salary: "))
da = 0.58 * basic_salary
ta = 0.30 * basic_salary
hra = 0.38 * basic_salary
gross_salary = basic_salary + da + ta + hra
print("Gross Salary is:", gross_salary)
```

```python
# 4.] Marks of 5 subjects are input to the keyboard. 
# Write a program to print a percentage of them.

sub1 = float(input("Enter marks of subject 1: "))
sub2 = float(input("Enter marks of subject 2: "))
sub3 = float(input("Enter marks of subject 3: "))
sub4 = float(input("Enter marks of subject 4: "))
sub5 = float(input("Enter marks of subject 5: "))

total_marks = sub1 + sub2 + sub3 + sub4 + sub5
percentage = (total_marks / 500) * 100
print("Percentage is:", percentage)
```

```python
# 5.] Accept any 5 digit number from the user, reverse it and check whether the original and reverse are 
# same or not.

num = int(input("Enter any 5 digit number: "))
reversed_num = 0
temp = num

while temp>0:
    digit = temp%10
    reversed_num = reversed_num*10 + digit
    temp = temp//10

if num == reversed_num:
    print("The original and reversed numbers are the same.")
else:
    print("The original and reversed numbers are different.")
```

```python
# 6.] Accept any year from the user and check whether it is a leap year or not.
# (A leap year is a year which is divisible by 4 but not by 100 and should be divisible by 400).

year = int(input("Enter any year: "))

if(year%4==0 and year%100!=0) or (year%400==0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
# 8.] Accept inner and outer radius of a ring and print 
# the perimeter and area of the ring 
# (Hint: perimeter = 2 π (a+b) , area = π (a*a-b*b) )

# import math
inner_radius = float(input("Enter inner radius of the ring: "))
outer_radius = float(input("Enter outer radius of the ring: "))

perimeter = 2 * 3.14 * (inner_radius + outer_radius)
area = 3.14 * (outer_radius**2 - inner_radius**2)    

print(f"Perimeter of the ring: {perimeter}")
print(f"Area of the ring: {area}")
```

```python
# 10.] Consider a room having one door and two windows 
# both of the same size. Accept dimensions of the room, 
# door and window. Print the area to be painted 
# (interior walls) and area to be whitewashed (roof).

room_length = float(input("Enter room length: "))
room_breadth = float(input("Enter room breadth: "))
room_height = float(input("Enter room height: "))

door_length = float(input("Enter door length: "))
door_breadth = float(input("Enter door breadth: "))

window_length = float(input("Enter window length: "))
window_breadth = float(input("Enter window breadth: "))

# Calculate area of interior walls
area_interior_walls = 2 * (room_length + room_breadth) * room_height

# Calculate area of door and windows
area_door = door_length * door_breadth
area_window = 2 * (window_length * window_breadth)

# Calculate area to be painted (interior walls)
area_to_be_painted = area_interior_walls - area_door - area_window

# Calculate area to be whitewashed (roof)
area_to_be_whitewashed = room_length * room_breadth

print(f"Area to be painted (interior walls): {area_to_be_painted}")
print(f"Area to be whitewashed (roof): {area_to_be_whitewashed}")
```

```python
#Write a python program to replace the last value of tuple in a list
# sample_list = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]
# Expected ouput: [(10, 20, 100), (40, 50, 100), (70, 80, 100)]

tuple_list = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]
new_value = 100

result = []
for t in tuple_list:
    new_tuple = t[:-1] + (new_value,)
    result.append(new_tuple)

print("Updated List:", result)
```

```python
# Convert a list with duplicate into a set and print all the unique values

sample_list = [10, 20, 30, 10, 20, 40, 50, 60, 40]
unique_values = set(sample_list)
print("Unique Values:", unique_values)
```

```python
# Given two set of numbers,find the common elements (Intersection)
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}  

common_elements = set1 & set2
print("Common Elements:", common_elements)
```

```python
# Given two sets, find the elements that are only in the first set

set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

only_in_set1 = set1 - set2
print("Elements only in the first set:", only_in_set1)
```

```python
# Given two sets, find the elements that are in either set but not both (symmetric difference)

set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

symmetric_difference = set1 ^ set2
print("Symmetric Difference:", symmetric_difference)
```

```python
# Check if one set is a subset of another set

set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

is_subset = set1.issubset(set2)
print("Is set1 a subset of set2?", is_subset)
```

```python
# Check whether given string is palindrome or not

str = "DAD"
reversed_str = str[::-1]

if str == reversed_str:
    print(f'"{str}" is a palindrome.')
else:
    print(f'"{str}" is not a palindrome.')

```

```python
# Reverse the string

str = "123456"
reversed_str = str[::-1]
print("Reversed String:", reversed_str)
```

```python
# Take input from user and check whether a given number is prime or not

num = int(input("Enter a number: "))

is_prime = True
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
else:
    is_prime = False

if is_prime:
    print("Prime number")
else:
    print("Not a prime number.")
```

```python
# Count the number of vowels in a given string

input_str = "TechGeekConnect"
vow = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
count = 0   

for char in input_str:
    if char in vow:
        count += 1

print("Number of vowels:", count)
```
